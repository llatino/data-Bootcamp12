# **Control flow**

* if
* for

```python
# if
score = 45
if score >= 82:
  print("Passed")
else:
  print("failed")
```

```python
# mutiple grades
score = 55

if score >= 80:
  print("A")
elif score >= 70:
  print("B")
elif score >= 60:
  print("C")
else:
  print("D")
```

```python
age: int = int(input("How old are u: "))
print(age)
```

```python
# mutiple condition
# nigth club
has_id = True
age_over_18 = False
vip_customer = True

if has_id and (age_over_18 or vip_customer):
  print("Welcome to the club!")
else:
  print("Please come next time!!")
```

```python
# for & while
items = ["egg", "milk", "bread"]

for item in items:
  if item == "milk":
    print("Milk CP Meji")
  else:
    print(item)
```

```python
# List comprehension
result = []
for item in items:
  result.append(item.upper())

print(result)
```

```python
[item.upper() for item in items]
```

```python
## While ทำซ้ำจนกว่าไม่เป็นจริง
while True:
  user_input = input("Play on: Y/N ")
  if user_input == "N":
    print("Goodbye!")
    break
  else:
    print("GOOOO!")
```

```python
# Count
count = 0
while count < 5:
  print("hi!")
  count += 1
```
